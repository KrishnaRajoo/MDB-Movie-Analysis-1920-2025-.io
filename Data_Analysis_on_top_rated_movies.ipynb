```python
def count_genre_plot (df):
    genre_counts = pd.DataFrame(df['GENERES'].value_counts())
    plt.figure(figsize=(10,6))
    sns.barplot( x = 'GENERES', y = 'count', hue = 'GENERES', data = genre_counts.head(30))
    plt.title('Count of Genres')
    plt.xlabel('Genres')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
def genre_sort(genre_df,col):
    df = genre_df
    if (col == 'VOTES'):
        df = df.groupby('GENERES')['VOTES'].sum()
    elif (col == 'RATING'):
        df = df.groupby('GENERES')['RATING'].mean()
    df = df.sort_values()
    df = pd.DataFrame(df)
    if (col == 'VOTES'):
        df['votes/million']=df['VOTES']/1000000
    df = df.reset_index()
    return df

def genre_plot(df,col):
    plot_data = df.tail(20)
    plt.figure(figsize=(10,6))
    if (col == 'VOTES'):
        sns.barplot( x = 'GENERES', y = 'votes/million', hue = 'GENERES', data = plot_data)
        plt.ylabel('Votes (in millions)')
        plt.title('Total Votes')
    elif (col == 'RATING'):
        sns.barplot( x = 'GENERES', y = 'RATING', hue = 'GENERES', data = plot_data)
        plt.ylabel('Average rating')
        plt.title('Average rating')
    plt.xlabel('Genres')
    plt.xticks(rotation=90)
    plt.show()
```

```python
df = pd.concat([s1,s2,s3,s4],axis=0)
df
```

```python
df['GENERES'] = df['GENERES'].str.replace(r'[\[\]"]', '', regex=True)
df['GENERES'] = df['GENERES'].str.replace('Musical', 'Music')
genre_set=set()
for genre in df['GENERES']:
    for item in genre.split(', '):
        genre_set.add(item)
genre_set
```

```python
data = pd.DataFrame(columns=['id', 'title', 'averageRating', 'numVotes', 'releaseYear', 'genre'])
for genre in genre_set:
    df2 = pd.DataFrame(columns=['id', 'title', 'averageRating', 'numVotes', 'releaseYear'])
    for i in range(0, len(df)):
        if genre in df.iloc[i,2]:
            df2 = pd.concat([df2, df.iloc[[i]]], ignore_index=True)
    df2['genre'] = genre
    data = pd.concat([df2, data], ignore_index=True)
```

```python
data
```

```python
data_grouped = data.groupby(['GENERES','YEAR'])['GENERES'].count()
data_grouped = pd.DataFrame(data_grouped)
data_grouped = data_grouped.rename(columns = {'GENERES':'GENERE_COUNT'})
data_grouped = data_grouped.reset_index()
data_grouped
```

```python
for genre in genre_set:
    temp = pd.DataFrame(columns=['releaseYear'])
    for i in range(0, len(data_grouped)):
        if genre in data_grouped.iloc[i,0]:
            temp = pd.concat([temp, data_grouped.iloc[[i]]], ignore_index=True)
    print(f"{genre}\n{temp}")
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x='YEAR',  y='GENERE_COUNT', data=data_grouped , hue='GENERES')
```

```python
plt.figure(figsize=(15,5))
genre_order = ['Horror', 'Crime', 'Drama', 'Action']
sns.lineplot(x='YEAR',  y='GENERE_COUNT', data=data_grouped , hue='GENERES', hue_order = genre_order)
plt.ylim(0,30)
```
